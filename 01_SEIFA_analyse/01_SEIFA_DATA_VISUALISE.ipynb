```python
df_adv = df_seifa_adv[['2021 Statistical Area Level 2  (SA2) 9-Digit Code',
       '2021 Statistical Area Level 2 (SA2) Name ',
       'Decile ranking of the SA2 within Australia']]

df_adv.columns = ['SA2_CODE', 'SA2_NAME', 'IRSAD_DECILE']

print('--- RENAMED COLUMNS DATA ---')
df_adv.head(5)
```

```python
df['IRSAD_DECILE'] = df_adv['IRSAD_DECILE']

df.head()
```

```python
df['GAP'] = df['IRSD_DECILE'] - df['IRSAD_DECILE']
df['GAP_ABS'] = abs(df['IRSD_DECILE'] - df['IRSAD_DECILE'])


df.head()
```

```python
# Get map file (.shp)
map_path = 'datas/SA2_2021_AUST_SHP_GDA2020/SA2_2021_AUST_GDA2020.shp'
gdf_map = gpd.read_file(map_path)

gdf_map

# Edit gdf_map 
gdf_map = gdf_map[['SA2_CODE21', 'GCC_NAME21','geometry']]
gdf_map.columns = ['SA2_CODE', 'GCC_NAME21', 'geometry']
gdf_map['SA2_CODE'] = gdf_map['SA2_CODE'].astype(str)

gdf_map
```

```python
df['SA2_CODE'] = df['SA2_CODE'].astype(str)
final_map = gdf_map.merge(df, on='SA2_CODE', how='left')

final_map.info()
```

```python
final_map.head()

final_map.to_csv('data_with_geometry.csv', sep=',', index=False)
```

```python
sydney = final_map[final_map['GCC_NAME21'] == 'Greater Sydney'].copy()

import folium

# base map 
m1 = folium.Map(location=[-33.8688, 151.2093], zoom_start=10.5, tiles='cartodbpositron')

# 1st layer : IRSD (disadvantage)
choropleth1 = folium.Choropleth(
    geo_data=sydney,
    name='IRSD (Disadvantage)', 
    data=sydney,
    columns=['SA2_CODE', 'IRSD_DECILE'],
    key_on='feature.properties.SA2_CODE',
    fill_color='RdYlGn',  # red(not good) -> green(good)
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='IRSD Decile'
).add_to(m1)

# 2nd layer : IRSAD (advantage)
choropleth1 = folium.Choropleth(
    geo_data=sydney,
    name='IRSAD (Advantage)', 
    data=sydney,
    columns=['SA2_CODE', 'IRSAD_DECILE'],
    key_on='feature.properties.SA2_CODE',
    fill_color='RdYlGn',  # red(not good) -> green(good)
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='IRSAD Decile'
).add_to(m1)

# add tool tip : score with mouse
folium.GeoJson(
    sydney,
    name='Tooltip',
    style_function=lambda x: {'color': 'transparent', 'fillColor': 'transparent', 'weight': 0},
    tooltip=folium.GeoJsonTooltip(
        fields=['SA2_NAME', 'IRSD_DECILE', 'IRSAD_DECILE', 'GAP'],
        aliases=['Suburb:', 'IRSD (Disadv):', 'IRSAD (Total):', 'Gap:'],
        localize=True
    )
).add_to(m1)

# add toggle button 
folium.LayerControl(collapsed=False).add_to(m1)

# SAVE THE MAP
m1.save('sydney_comparison_map.html')
print("Toggle Map is saved as 'sydney_comparison_map.html'! ")
```

```python
# CREATE BASE MAP 
m2 = folium.Map(location=[-33.8688, 151.2093], zoom_start=10.5, tiles='cartodbpositron')

# VISUALISE THE GAP 
folium.Choropleth(
    geo_data=sydney,
    name='Gap Analysis',
    data=sydney,
    columns=['SA2_CODE', 'GAP'],
    key_on='feature.properties.SA2_CODE',
    fill_color='RdBu_r', 
    fill_opacity=0.8,
    line_opacity=0.2,
    legend_name='Gap (IRSAD - IRSD)',
    bins=list(range(-10, 11)) 
).add_to(m2)

# 3. 툴팁 (Gap 설명 추가)
folium.GeoJson(
    sydney,
    name='Gap Tooltip',
    style_function=lambda x: {'color': 'transparent', 'fillColor': 'transparent', 'weight': 0},
    tooltip=folium.GeoJsonTooltip(
        fields=['SA2_NAME', 'GAP', 'IRSD_DECILE', 'IRSAD_DECILE'],
        aliases=['Suburb:', 'Gap Score:', 'IRSD:', 'IRSAD:'],
        localize=True
    )
).add_to(m2)

# 4. 저장
m2.save('sydney_gap_map.html')
print("격차 지도가 'sydney_gap_map.html'로 저장되었습니다!")
```
